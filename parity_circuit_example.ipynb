```python
#Notes: I had a problem with measurements so I used statevector and measured all the states simulatiously,anyways what we care about is the measurement of
#the second qubit,I run 2 blocks of code.The first one runs for the function f(0)=f(2)=f(3)=0 and f(1)=1.We measure 2 times,one after the first query
#and one after the second query.Looking at the second qubit we get 1 and 0,as we proved theoretically the parity is 1.
#In the next block we do the same but now the function is the constant f(0)=f(1)=f(2)=f(3),we see the values of the second qubit be 0 and again 0 so parity 0
#So these 2 test experimentally the idea of the algorithm.



from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# Define the circuit
qc = QuantumCircuit(3, 1)

# Convert the OpenQASM code to Qiskit
qc.reset(0)
qc.h(1)
qc.x(2)
qc.h(2)
qc.barrier()
qc.cx(1, 2)
qc.barrier()
qc.h(1)
qc.barrier()
# Simulate the statevector
state = Statevector.from_instruction(qc)

# Sample measurement outcomes (measuring only qubit 1, mapped to classical bit 0)
num_shots = 1024  # Number of measurement shots
counts = state.sample_counts(shots=1024)

# Display the circuit
print("Quantum Circuit:")
print(qc.draw(output='text'))

# Display the results
print("\nFinal Statevector:")
print(state)

print("\nMeasurement Counts:")
print(counts)

qc.reset(2)
qc.x(2)
qc.x(0)
qc.reset(1)
qc.h(1)
qc.h(2)
qc.barrier()
qc.barrier()
qc.h(1)
# Simulate the statevector
state = Statevector.from_instruction(qc)

# Sample measurement outcomes (measuring only qubit 1, mapped to classical bit 0)
num_shots = 1024  # Number of measurement shots
counts = state.sample_counts(shots=1024)

# Display the circuit
print("Quantum Circuit:")
print(qc.draw(output='text'))

# Display the results
print("\nFinal Statevector:")
print(state)

print("\nMeasurement Counts:")
print(counts)
```
